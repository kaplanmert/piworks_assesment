**Kütüphanelerin ve verilerin yüklenmesi**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/sample_data/country_vaccination_stats.csv')


```

```python
df.head()
```

**NaN verilerin sayısını öğrenme**

```python
df.isna().sum()
```

**Günlük aşı sayılarındaki NaN verilerin yerine, alakalı ülkelerdeki minimum verileri koymak için, ülke ülke minimum aşı sayılarının olduğu bir veri seti yapildi.**

```python
min_country = df.groupby('country').agg({'daily_vaccinations':'min'})
```

```python
min_country
```

**Ana veri setimizi, günlük aşı verilerine göre sıraladığımızda sonda kalacak 60 veri sadece NaN verilerini içerecektir. Yukarıdaki minimum değerlerle değiştirmek amacıyla NaN verilerinin olduğu bir veri seti yapildi.**

```python
a = df.sort_values(by='daily_vaccinations').tail(60)
pd.DataFrame(a)
```

**Önce veri setinin indexleri sıfırlandı.**

**Her ülkeden birer kayıp değerimiz olduğu için bunları index index eşitleyerek Kayıp değerlerden kurtulundu.**

```python
a = a.reset_index()
for i in range(60):
       a['daily_vaccinations'][i] = min_country['daily_vaccinations'][i]
```

```python
a
```

**Kuveyt değeri olan NaN, 0 olarak düzeltildi.**

```python
a = a.fillna(0)
```

**Ana verisetimizdeki NaN verileri önce veri setinden çıkardık daha sonrasında,a verisetimizle birleştirdik ve indexleri resetledik.**

```python
df = df.dropna()
df = pd.concat([df,a])
df = df.reset_index()
```

**5. Soru : Ülkelerin günlük aşı sayılarına göre gruplanması. Görüldüğü üzere verilere göre en çok günlük aşıyı United States, China ve United Kingdom yapmıştır.**

```python
daily_vac_by_country = df.groupby('country').agg({'daily_vaccinations':'sum'})
daily_vac_by_country.sort_values(by = 'daily_vaccinations', ascending = False)
```
